# This code performs classification on a dataset using the XGBoost algorithm.
## Importing libraries:
pandas for data manipulation, sklearn for preprocessing and evaluation metrics, and xgboost for the XGBoost classifier.

```python
import pandas as pd
voice=pd.read_csv('voice.csv')
voice.head()
```

### Provides statistical information about the dataset.

```python
voice.describe()
```

The code imports the preprocessing module from sklearn to perform data preprocessing tasks.

# # LabelEncoder()
is used to encode the labels in the "label" column of the voice dataframe into numerical values.

# # le.classes_
displays the distinct classes in the encoded labels.

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
voice["label"] = le.fit_transform(voice["label"])
le.classes_
```

# # MinMaxScaler() 
applied to scale the numerical features of the voice dataframe to a range between 0 and 1.

```python
voice[:]=preprocessing.MinMaxScaler().fit_transform(voice)
voice.head()
```


# Splitting the training and test data:

```python
from sklearn.model_selection import train_test_split
import xgboost
```

# # train_test_split()
is used to split the preprocessed dataset into training and test sets, with a test size of 0.3 (30% of the data is used for testing).

```python
train, test = train_test_split(voice, test_size=0.3)
```

```python
train.head()
```

# # classify()
takes a model, training data, and test data as input.

The model is trained using the fit() method on the training data.

# # classification_report()
is used to generate a report containing precision, recall, F1-score, and support for each class.

The function displays the classification report.

```python
def classify(model,x_train,y_train,x_test,y_test):
    from sklearn.metrics import classification_report
    target_names = ['female', 'male']
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(classification_report(y_test, y_pred, target_names=target_names, digits=4))
```

## Comparing the performance of the machine learning model on the whole dataset versus using only the 5 variables

The variables x_train, y_train, x_test, and y_test are used to store both the input features and output labels for both the training and test data.


```python
x_train = train.iloc[:, :-1]
y_train = train["label"]
x_test = test.iloc[:, :-1]
y_test = test["label"]

model = xgboost.XGBClassifier()
classify(model,x_train,y_train,x_test,y_test)
```

```python
x_train2 = train[["meanfun","sd","median","Q25","Q75"]]
y_train2 = train["label"]
x_test2 = test[["meanfun","sd","median","Q25","Q75"]]
y_test2 = test["label"]

model = xgboost.XGBClassifier()
classify(model,x_train2,y_train2,x_test2,y_test2)
```

### The model trained on the whole dataset slightly outperforms the model trained with only 5 variables in terms of accuracy and F1-scores.
### However, the difference in performance between the two models is relatively small, suggesting that the 5 variables capture a significant portion of the information needed for classification.
### Using only the 5 variables may provide a more efficient and simpler model without sacrificing much in terms of performance.
